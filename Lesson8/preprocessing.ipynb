```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('Sales/Sales_1.csv',';')
number = 2
while number <= 20:
    df_tmp = pd.read_csv('Sales/Sales_' + str(number) + '.csv',';')
    df = pd.concat([df, df_tmp], axis=0, ignore_index=True)
    number = number + 1
df.info()
```

```python
df.drop('Unnamed: 10', axis=1, inplace=True)
df.head()
```

```python
df['Год'] = df['Год'].str.replace('\s', '')
df['ТоварКод'] = df['ТоварКод'].str.replace('\s', '')
df['ПроизводительКод'] = df['ПроизводительКод'].str.replace('\s', '')
df['Количество'] = df['Количество'].str.replace('\s', '')
df['СуммаПродажи'] = df['СуммаПродажи'].str.replace('\s', '')
df['Себестоимость'] = df['Себестоимость'].str.replace('\s', '')
```

```python
df['Количество'] = df['Количество'].str.replace(',', '.')
df['СуммаПродажи'] = df['СуммаПродажи'].str.replace(',', '.')
df['Себестоимость'] = df['Себестоимость'].str.replace(',', '.')
```

```python
df['Количество'] = df['Количество'].astype('float')
df['СуммаПродажи'] = df['СуммаПродажи'].astype('float')
df['Себестоимость'] = df['Себестоимость'].astype('float')
df['Год'] = df['Год'].astype('int')
```
